#### Chart - 3 Distribution of Handling Time

````python
df['Handling_Time'] = (df['issue_responded'] - df['Issue_reported at']).dt.total_seconds() / 60
plt.figure(figsize=(8, 5))
sns.histplot(df['Handling_Time'].dropna(), bins=50, kde=True, color='blue')
plt.title('Distribution of Handling Time (in minutes)')
plt.xlabel('Handling Time (minutes)')
plt.ylabel('Frequency')
plt.show()
````

##### 1. Why did you pick the specific chart?

A histogram with KDE (Kernel Density Estimation) helps analyze the distribution of handling time, which is critical for understanding service efficiency.

This visualization helps:

 Assess if handling times follow a normal pattern or need process optimization.
